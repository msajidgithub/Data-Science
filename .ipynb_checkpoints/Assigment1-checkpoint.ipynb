```python
#question 7: take input of 5 subject mark
english = int(input("Enter English: "))
urdu = int(input("Enter Urdu: "))
math = int(input("Enter Math: "))
islamiat = int(input("Enter Islamiat: "))
computer = int(input("Enter Com12puter: "))
marks = [english, urdu, math, islamiat, computer]
total = english + urdu + math + islamiat + computer
percentage = total / 5
print("Marks List:", marks)
if (0 <= english <= 100 and 0 <= math <= 100 and 0 <= urdu <= 100 and 0 <= islamiat <= 100 and 0 <= computer <= 100):
    total = english + math + urdu + islamiat + computer
    percentage = total / 5
    print("Total Marks:", total)
    print("Percentage:", percentage)

    if percentage >= 80:
        print("Grade A")
    elif percentage >= 60:
        print("Grade B")
    elif percentage >= 40:
        print("Grade C")
    else:
        print("Fail")

else:
    print("Invalid marks! Each subject must be between 0 and 100.")
```
