## ***Problem Statement:***

**In today’s job market, professionals face significant challenges when it comes to career growth and job satisfaction. While there is an increasing demand for skilled workers, especially in fields like data science, many employees remain stuck in middle level or entry level roles with limited opportunities for advancement into senior or executive positions. There are also noticeable salary differences between small, medium, and large companies. Small companies often offer much lower wages, making it harder for them to compete with larger organizations that can provide more attractive compensation packages. This mismatch between available talent, required skills, and company offerings complicates the job market further.**

**The dataset appears to contain various columns with information about employees, their work details, salaries, job titles, and other related attributes. Here's a quick overview of the columns:**

`work_year:` The year the data was recorded.

`experience_leve:` The experience level of the employee (e.g., MI = Mid-level, SE = Senior-level).

`employment_type:` The type of employment (e.g., FT = Full-time).

`job_title:` The specific job title of the employee.

`salary:` The salary of the employee in the respective currency.

`salary_currency:` The currency of the salary (e.g., EUR, USD, GBP).

`salary_in_usd:` The salary converted into USD.

`employee_residence:` The country where the employee resides.

`remote_ratio:` The ratio of remote work (percentage of time working remotely).

`company_location:` The location of the company.

`company_size:` The size of the company (e.g., L = Large, M = Medium, S = Small).

## **Exploring Workforce Dynamics:**


Examining the distribution of work experience across different years, shedding light on the average salaries for each year. The analysis revealed trends in salary progression over time, offering valuable insights for career planning and salary negotiations.

```python
df1 = df.groupby('work_year')['salary_in_usd'].mean().round(2)
df1
```

```python
df1.index
```

```python
df1.values
```

```python
data = {
    'work_year': df1.index,
    'average_salary': df1.values
}
df1 = pd.DataFrame(data)
df1
```

```python
df1['average_salary'] = (df1['average_salary']/1000).round(2)
df1
```

```python
df1.plot(kind = 'bar', x = 'work_year', y = 'average_salary')
```

```python
dir(pd)
```

```python
ax = df1.plot(kind = 'bar', x = 'work_year', y = 'average_salary', legend = True)
ax.bar_label(ax.containers[0], labels = df1['average_salary'].map('{:.2f}K'.format))
ax.legend(['Average Salary'], loc = 'upper left')
plt.xlabel('Work Year')
plt.ylabel('Average Salary')
plt.title('Average salaries by years', color = 'red')
plt.show()
```

***The average salary showed a consistent upward trend from 95.81K in 2020 to 124.52K in 2022, with a notable increase in 2022, indicating growing demand for skilled professionals or changes in industry compensation strategies.***

## **Salary Distribution by Company Size:**

Analysis investigated the distribution of salaries across different company sizes, providing insights into how salary scales vary based on the size of the company. By visualizing the distribution of salaries, we gained insights into salary trends and disparities across different company sizes.

```python
df5 = df[["salary_in_usd", "company_size"]]
df5
```

```python
df5[df5['company_size'] == 'S']
```

```python
S = df5[df5['company_size'] == 'S']
M = df5[df5['company_size'] == 'M']
L = df5[df5['company_size'] == "L"]

labels = ['Small','Medium', 'Large']
sal_mean = [S['salary_in_usd'].mean(), M["salary_in_usd"].mean(), L["salary_in_usd"].mean()]
sal_mean
```

```python
S
```

```python
for x in sal_mean:
    print(x)
```

```python
label_change = np.round([x for x in sal_mean], 1)
# label_change
label_change = list(map(str, label_change))
label_change = [x + '$' for x in label_change]
label_change
```

```python
z = plt.bar(labels, sal_mean)
plt.bar_label(z,label_change, alpha = 0.8)
plt.title('Distribution of salary by comapny size')
plt.xlabel('Size of company')
plt.ylabel('Salary in usd')
plt.show()
```

```python
sns.histplot(S['salary_in_usd'], label = 'Small', kde = True)
sns.histplot(M['salary_in_usd'], label = 'Medium', kde = True)
sns.histplot(L['salary_in_usd'], label = 'Large', kde = True)
plt.title('Distribution salary by comapny size')
plt.xlabel('salary in usd')
plt.show()
```

***Employees in larger companies generally earn higher salaries, followed by those in medium sized companies, with small companies offering the lowest salaries. This pattern is likely due to the greater financial resources and compensation capabilities of larger organizations.***
