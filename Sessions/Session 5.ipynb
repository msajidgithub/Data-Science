```python
quiz = [
    {
    "S No." : 1,
    "question" : "What is your country name",
    "options": ["Pakistan","India","China","Iran"],
    "answer": "Pakistan"
    },
    {
    "S No." : 2,
    "question" : "What is your city name",
    "options": ["Karachi","Islamabad","Quetta","Sukkur"],
    "answer": "Pakistan"
           
    },
    {
    "S No." : 3,
    "question" : "What is your  name",
    "options": ["Ali","Amjad","Adnan","Anjum"],
    "answer": "Ali"
    },
    {
    "S No." : 4,
    "question" : "What is your id",
    "options": ["123","456","789","345"],
    "answer": "456"
    }
       ]
score = 0
for quizApp in quiz:
    data = str(quizApp["S No."]) + " " + quizApp["question"] + "\n" + quizApp["options"][0] + " " + quizApp["options"][1] + " " + quizApp["options"][2] + " " + quizApp["options"][3]
    ans = input(data)
    if ans == quizApp["answer"]:
        score += 1
print("Your Score for the quiz is", score)



```
